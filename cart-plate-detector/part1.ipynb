### Generate an image with a symbol

```python
# import string
from matplotlib import pyplot as plt
from PIL import Image,ImageDraw,ImageFont

# print(string.ascii_lowercase)
# print(string.ascii_uppercase)
# print(string.digits)

C_W = 75 
C_H = 100 


chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789-."

def generateCharacter(char,size,color,bgColor,fontName):
    # GENERATE AN IMAGE WITH A SINGLE CHARACTER
   
    img = Image.new(mode="L",  size= size , color= bgColor)
    draw = ImageDraw.Draw(img)

    font = ImageFont.truetype(f"./fonts/{fontName}.ttf", int(size[1] * 0.8 ))
   

    draw.text((0,0), char, fill= color, font= font)

    return img
```

```python
img1= generateCharacter("y",(C_W,C_H), 255, 0,"arial")

plt.figure()
plt.imshow(img1)
img1= generateCharacter("y",(C_W,C_H), 255, 0,"times")

plt.figure()
plt.imshow(img1)
img1= generateCharacter("y",(C_W,C_H), 255, 0,"courier")

plt.figure()
plt.imshow(img1)
```
