### A function derivative represents its "speed of Change"


```python
def linear(x):
    w = 1
    b = 0
    return w * x + b

def linear_prime(x):
    return 1

def sigmoid(x):
    e = 2.718281828459045
    return 1/ (1+2.7 ** -x)

def sigmoid_prime_approx(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
import matplotlib.pyplot as plt

X = [-10, -5, -4, -3, -2, -1, -0.5, 0, 0.5, 1, 2, 3, 4, 5, 10]

Y = []
Yp = []

for x in X:
    Y.append(linear(x))
    Yp.append(linear_prime(x))
plt.plot(X,Y)
plt.plot(X,Yp)
plt.show()
```
