## Why MSE (Mean SQUARE Error)?
* to obtain a local "minimum" of error

```python
def le(y,yp):
    return y - yp

def se(y, yp):
    return(y - yp) ** 2
```

```python
import matplotlib.pyplot as plt

Y = [-10, -5, -4, -3, -2, -1, -0.5, 0, 0.5, 1, 2, 3, 4, 5, 10]

El = []
Es = []

for y in Y:
    El.append(le(1,y))
    Es.append(se(1,y))

plt.plot(Y,El)
plt.plot(Y,Es)
plt.show()
```
