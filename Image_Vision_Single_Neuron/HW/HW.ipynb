## LET'S train a neuron to "detect" positive numbers in the first column.
* the numerical values will always be in the range -5 .. 5
<br>
Build and train a model with a single neuron, which will be able to detect if the matrix contains positive numbers in the first column.
<br>

* The model must give a value higher than 0.7 at the output when the first column in the matrix has ALL positive numbers and values lower than 0.7 when the first column has either negative or zero numbers
* So for the  matrix the result of the neuron would be > 0.7 (ideally 1)

```python
# bach of TRAINING data
Xtrain = [
    [ # -----> 1
    [5, -1,  0],
    [2,  0, -5],
    [1,  1, -5],
    ],
    [ # ----->  0
    [1, 1,  0],
    [-2, -1, -5],
    [-5,  1, 0],
    ],
    [ # ----->  0
    [ 2,  1, -5],
    [-2,  0,  0],
    [-1,  1, 0],
    ],
    [ #  ---->  0
    [ -5,  0, 1],
    [ -1,  0, 0],
    [ -2,  1, 0]
    ],
    [ # -----> 0
    [ -1, 2,  1],
    [  1, 0,  2],
    [  2, 0,  2],
    ],
    [ # -----> 0
    [-2,  0,  0],
    [ 1,  2, -1],
    [ 1,  1, -1]
    ],
    [ # -----> 0
    [  1, 0,  0],
    [ -1, 0,  5],
    [  1,-1, -1]
    ]
]



Ytrain = [
    1,
    0,
    0,
    0,
    0,
    0,
    0
]

# bach of TESTING data
Xtest = [
    [ 
    [-2, -1,  0],
    [-2,  0, -5],
    [ 1,  1, -5],
    ],
    [ 
    [-5, -1,  0],
    [-2,  0, -5],
    [ 1,  1,  1],
    ],
    [
    [ 2, -1,  0],
    [ 5,  0, -5],
    [-2,  1, -5],
    ]
]
Ytest = [
    0,
    0,
    0
    ]


```

```python
# x - single value
# X - collections

class LinearNeuronDetect:

    def __init__(self):
        self.weights = [0 for i in range(9)]
        self.bias = 0
    
    # def stepActivation(x):
    #     return 1 if x >=1 else 0
    
    
    def sigmoidActivation(self, x ):
        e = 2.71828182845905
        return 1/ ( 1 + e ** - x ) 
    
    def forward(self, X):
        y = self.bias
        for i in range(9):
            y += X[i]* self.weights[i]
        
        y = self.sigmoidActivation(y)
        
        return y

    def __str__(self):
        out = ""
        for i in range(9):
            if i  % 3 == 0:
                out += "\n"
            
            out += f"{self.weights[i]:10.5f} "

        out += "\n"    
        out += f"{self.bias:10.5f}"
        return out

def loss(Y, Yp):
    return Y - Yp
    
def reshapeMatrix(X):
    Xt = []
    for i in range(3):
        for j in range(3):
            Xt.append( X[i][j])
    return Xt
    
```

### Neuron traning

```python
neuron = LinearNeuronDetect()

lr =0.01

for epoch in range(10000):
    errors = []
    for i in range(7):

        # FORWARD  PASS  
        xt = Xtrain[i]
        yt = Ytrain[i]

        xt = reshapeMatrix(xt)
        yp = neuron.forward(xt)

        # ESTIMATE LOSS
        e = loss(yt, yp)
        
        # BACKPROPAGATION / OPTIMIZATION
        dBias = lr * e
        neuron.bias += dBias

        for j in range(9):
            dweight = lr * e * xt[j]
            neuron.weights[j] += dweight
        errors.append(e)
    # epoch stats
    if epoch % 1000 == 0:
        me = abs(sum(errors)/7)
        print("--------------------------------------------")
        print(f"epoch:{epoch}")    
        print(f"loss:{me:10.5f}")
        #print(neuron)

```

### TESTING

```python
X = Xtrain + Xtest
Y = Ytrain + Ytest
for i in range(9):
        xt = X[i]
        yt = Y[i]
        xt = reshapeMatrix(xt)
        yp = neuron.forward(xt)
        e = loss(yt, yp)
        # print(f" loss: {e:10.5f} Yp{yp: 10.5f}")
```
